```python
import pandas as pd

df = pd.read_csv("istherecorrelation.csv", sep=";")
df = df.replace(",", ".", regex=True)
df.iloc[:,1] = df.iloc[:,1].astype("float64")
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

years = df["Year"]

fig, ax = plt.subplots(figsize=(5, 4), dpi=300)
plt.plot(years, df["WO [x1000]"], color="black")
ax.set_ylabel("WO students (x1000)", color="black")

ax2 = ax.twinx()
ax2.plot(years, df["NL Beer consumption [x1000 hectoliter]"], color="red")
ax2.set_ylabel("NL Beer consumption in hectoliter (x1000)", color="red", rotation=-90, labelpad=14)
plt.show()
```
