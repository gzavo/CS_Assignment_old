```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("istherecorrelation.csv", delimiter = ";")
```

```python
data = data.rename(columns={"Year":"year","WO [x1000]":"WO", "NL Beer consumption [x1000 hectoliter]":"Consumption"})
```

```python
data["WO"] = data['WO'].str.replace(',','.')
data["WO"] = data["WO"].astype(float)
```

```python
R=np.corrcoef(data.WO, data.Consumption)[0][1]
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(data.year, data.WO, color="red", marker="o")
ax.set_title(label=f"Correlation between students and beer consumption\n R={round(R, 3)}")
# set x-axis label
ax.set_xlabel("year",fontsize=14)
# set y-axis label
ax.set_ylabel("University Students",color="red",fontsize=14)
# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(data.year, data["Consumption"],color="blue",marker="o")
ax2.set_ylabel("Consumption",color="blue",fontsize=14)
plt.savefig("consumptionGraph.jpg", 
            format="jpeg",
            dpi=300,
            bbox_inches='tight')
```
