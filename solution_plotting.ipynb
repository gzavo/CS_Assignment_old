```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
eg = pd.read_csv('istherecorrelation.csv', sep = ';')
```

```python
print(eg)
eg.columns
```

```python
x = eg['WO [x1000]']
y = eg['NL Beer consumption [x1000 hectoliter]']
```

```python
fig = plt.figure(figsize = (13, 13))
plt.scatter(x, y, color = 'darkred') 
plt.title('A plot to show the correlation between WO [x1000] and NL Beer consumption [x1000 hectoliter]', size = 15)
plt.xlabel('WO [x1000]', size = 13)
plt.ylabel('NL Beer consumption [x1000 hectoliter]', size = 13)
plt.grid(True)
plt.savefig("amazing.png", dpi = 300)
plt.show()
plt.close()
```
