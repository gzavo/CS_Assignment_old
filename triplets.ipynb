```python
import pandas as pd
import numpy as np 
import networkx as nx 
import io 
from sklearn.feature_selection import mutual_info_classif
```

```python
def cal_mi(df_data):
  num_var = df_data.shape[1]
  mi = pd.DataFrame(np.zeros(shape = (num_var, num_var)))
  
  for rv_f in range(num_var):
      for rv_t in range(num_var):
        var1_ser = df_data.iloc[:, rv_f]
        var2_ser = df_data.iloc[:, rv_t]

        mi.iloc[rv_f, rv_t] = mutual_info_classif(np.transpose(var1_ser.to_numpy()).reshape(-1, 1),
                                                var2_ser.to_numpy(),
                                                discrete_features=True)
  return mi
  
```

```python
def cond_ent_mat(mi, ch_method):
    num_var = len(mi)
    cond_ent_matrix = pd.DataFrame(np.zeros(shape = (num_var, num_var)))
    for ii in range(num_var):
        for jj in range(num_var):
  
            if ii < jj: # upper triangle
                cond_ent_1 = mi.iloc[ii, ii] - mi.iloc[ii, jj]
                cond_ent_2 = mi.iloc[jj, jj] - mi.iloc[ii, jj]
            else:
                continue
            
            if ch_method == 'min':
                cond_ent_matrix.iloc[ii, jj] = np.min([cond_ent_1, cond_ent_2])
                cond_ent_matrix.iloc[jj, ii] = np.min([cond_ent_1, cond_ent_2])
            elif ch_method == 'mean':
                cond_ent_matrix.iloc[ii, jj] = np.mean([cond_ent_1, cond_ent_2])
                cond_ent_matrix.iloc[jj, ii] = np.mean([cond_ent_1, cond_ent_2])
            elif ch_method == 'directed':
                cond_ent_matrix.iloc[ii, jj] = cond_ent_2
                cond_ent_matrix.iloc[jj, ii] = cond_ent_1

    return cond_ent_matrix
```

```python
def net_mat_threshold(mat, top_per):
  #Reshapes the input matrix 'mat' into a 1-dimensional array and sorts it with the highest value first
  mat_new = mat.copy()
  reshaped = mat_new.values.flatten()

  mat_sorted = sorted(reshaped, reverse=True)
  # Theshold value is calculated as the "round(mat.size * top_per) - 1" 
  threshold = mat_sorted[round(mat_new.size * top_per) - 1]
  return threshold
```

```python
def list2tuple_in_list(list_in_list):
  tuple_in_list = []
  for list_ele in list_in_list:
      tuple_in_list.append(tuple(list_ele))
  return tuple_in_list
```

```python
G = nx.from_numpy_array(adj_mat1.to_numpy())
```

```python
import itertools
```

```python
three_cliques = list(itertools.takewhile(lambda x: len(x) <= 3, nx.enumerate_all_cliques(G)))
```

```python
only_three_cliques = [i for i in three_cliques if len(i) == 3]
```

```python
triadic_cliques_tuple = list2tuple_in_list(only_three_cliques)
```

```python
def turn_to_int(strtup):
  ls = []
  for i in strtup:
     ls.append(int(i))
  tup = tuple(ls)
  return tup
```

```python
tri_clq_df["combint"] = tri_clq_df["comb"].apply(lambda x: turn_to_int(x))
```

```python
def ave_ch(ch_matrix, tri_clqs):
  tri_clqs_ch = tri_clqs.copy()
  tri_clqs_ch["ave_ch"] = 0 

  for row in range(len(tri_clqs)):
    tup = tri_clqs["combint"][row]
    var1, var2, var3 = tup

    mean_ch1 = ch_matrix.iloc[var1, var2]
    mean_ch2 = ch_matrix.iloc[var1, var3]
    mean_ch3 = ch_matrix.iloc[var2, var3]
  
    tri_clqs_ch.loc[row, "ave_ch"] = (mean_ch1 + mean_ch2 + mean_ch3)/3

  return tri_clqs_ch
```

```python
tri_clq_ch_df = ave_ch(ch_mat, tri_clq_df)
```
